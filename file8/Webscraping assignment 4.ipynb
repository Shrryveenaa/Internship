Scrape the details most watched tv series of all time from imdb.com. 
Url = https://www.imdb.com/list/ls095964455/ You 
have to find the following details: 
A) Name 
B) Year span 
C) Genre 
D) Run time 
E) Ratings 
F) Votes

```python
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
name=[]
yearSpan=[]
genre=[]
runTime=[]
ratings=[]
votes=[]

try:
    nameTag=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')
    name.extend([i.text for i in nameTag])
except:
    name.append("-")
    
try:
    yearSpantag=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
    yearSpan.extend([i.text for i in yearSpantag])
except:
    yearSpan.append("-")
    
try:
    genreTag=driver.find_elements(By.XPATH,'//span[@class="genre"]')
    genre.extend([i.text for i in genreTag])
except:
    genre.append("-")
    
try:
    runTimeTag=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    runTime.extend([i.text for i in runTimeTag])
except:
    runTime.append("-")

try:
    ratingsTag=driver.find_elements(By.XPATH,'//span[@class="ipl-rating-star__rating"]')
    ratings.extend([i.text for i in ratingsTag[::23]])
except:
    ratings.append("-")
    
try:
    votesTag=driver.find_elements(By.XPATH,'//span[@name="nv"]')
    votes.extend([i.text for i in votesTag])
except:
    votes.append("-")
```

```python
df=pd.DataFrame()
df['Name']=name
df['YearSpan']=yearSpan
df['Ratings']=ratings
df['RunTime']=runTime
df['Genre']=genre
df['Votes']=votes

df
```
