product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for 
guitars. 
2. In the above question, now scrape the following details of each product listed in first 3 pages of your search

scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings("ignore")
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://www.amazon.in/")
product_text=driver.find_element(By.ID,"twotabsearchtextbox")
product= input("enter the product name: ")
product_text.send_keys(product)
submit=driver.find_element(By.XPATH,"//input[@type='submit']")
submit.click()
```

```python
#Q2
product_url = []
url= driver.find_elements(By.XPATH,"//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")


for i in url:
    product_url.append(i.get_attribute('href'))
len(product_url)
```

```python
price=[]
rep=[]
brand=[]
delivery=[]
availablity=[]
product_name=[]
```

```python
#by using try and except we can scrap all the details by clicking on product url

for i in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        prices = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[2]/div[1]/div/span[1]/span[2]/span[2]")
        price.append(prices.text.replace('\n00',""))
    except NoSuchElementException:
        price.append('-')
        
    try:
        replacement = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span")
        rep.append(replacement.text)
    except NoSuchElementException:
        rep.append('-')
        
    try:
        brands = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span")
        brand.append(brands.text)
    except NoSuchElementException:
        brand.append('-')
        
    try:
        deli = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[9]/div[1]/div/div/div/span/a[1]")
        delivery.append(deli.text)
    except NoSuchElementException:
        delivery.append('Not Present')
        
    try:
        avail = driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[4]/div/div[1]/span")
        availablity.append(avail.text)
    except NoSuchElementException:
        availablity.append('-')
        
    try:
        name=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span")
        prod_name=name.text
        product_name.append(prod_name)
        if(prod_name == 0):
            try:
                name1=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1")
                product_name.append(name1.text)
            except NoSuchElementException:
                product_name.append('-')
                
    except:
        product_name.append("-")
        
driver.close()
        
    
```

```python
product = pd.DataFrame({'Product Name':product_name,'Brand':brand,'Price':price,'Availability':availablity,'Delivery':delivery,'Replacement/Exchange':rep,'URL':product_url})
product
```

6. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
# Activating the chrome browser
driver=webdriver.Chrome() 
time.sleep(2)
```

```python
#Opening the specified url
url = "https://www.digit.in/"
driver.get(url)
time.sleep(3)
#searching for best laptop
best_gam_lap = driver.find_element(By.XPATH,"//div[@class='listing_container']//ul//li[9]").click()
time.sleep(4)
```

```python
#Creating empty lists
lap_name = []
ope_sys = []
display = []
processor = []
memory = []
weight = []
dimensions = []
price = []
```

```python
# Scraping the data of laptop names
name_tags = driver.find_elements(By.XPATH,"//table[@id='summtable']//tr//td[1]")
for name in name_tags:
    lap_name.append(name.text)
    
    
#Scraping the data of operating system
try:
    os_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[3]//td[3]")
    for os in os_tags:
        ope_sys.append(os.text)
except NoSuchElementException:
    pass


#Scraping data of display
try:
    disp_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[4]//td[3]")
    for disp in disp_tags:
        display.append(disp.text)
except NoSuchElementException:
    pass


#Scraping data of Processor
try:
    pro_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[5]//td[3]")
    for pro in pro_tags:
        processor.append(pro.text)
except NoSuchElementException:
    pass


#Scraping data of memory
try:
    memo_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[6]//td[3]")
    for memo in memo_tags:
        memory.append(memo.text)
except NoSuchElementException:
    pass
#Scraping data of weight
try:
    wgt_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[7]//td[3]")
    for wgt in wgt_tags:
        weight.append(wgt.text)
except NoSuchElementException:
    pass


#Scraping data of dimensions
try:
    dim_tags = driver.find_elements(By.XPATH,"//div[@class='Spcs-details']//tr[8]//td[3]")
    for dim in dim_tags:
        dimensions.append(dim.text)
except NoSuchElementException:
    pass

#Scraping data of price
try:
    pri_tags = driver.find_elements(By.XPATH,"//td[@class='smprice']")
    for pri in pri_tags:
        price.append(pri.text.replace('₹','Rs '))
except NoSuchElementException:
    pass
```

```python
#DATA FRAMEING
Gaming_Laptop=pd.DataFrame({})
Gaming_Laptop['Laptop Name'] = lap_name
Gaming_Laptop['Operating system'] = ope_sys
Gaming_Laptop['Display'] = display
Gaming_Laptop['Processor'] = processor
Gaming_Laptop['Memory'] = memory
Gaming_Laptop['Weight'] = weight
Gaming_Laptop['Dimensions'] = dimensions
Gaming_Laptop['Price'] = price
```

“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 

```python
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Creating an instance of webdriver for google chrome
driver=webdriver.Chrome()
driver.maximize_window()
# Using webdriver we'll now open the flipkart website in chrome
driver.get("https://www.forbes.com/billionaires/")
driver.implicitly_wait(5)
```

```python
rank=[]
name=[]
net=[]
country=[]
source=[]
industry=[]


forbes=driver.find_elements(By.XPATH,"//div[@class = 'rank']")
for i in forbes:
    rank.append(i.text)
    
    
    
forbes=driver.find_elements(By.XPATH,"//div[@class = 'Name']")
for i in forbes:
    name.append(i.text)
    
    
    
forbes=driver.find_elements(By.XPATH,"//div[@class = 'netWorth']")
for i in forbes:
    net.append(i.text)
    
    
    
forbes=driver.find_elements(By.XPATH,"//div[@class = 'country']")
for i in forbes:
    country.append(i.text)
    
    
    
forbes=driver.find_elements(By.XPATH,'//div[@class="source"]')
for i in forbes:
    source.append(i.text)
    
    
    
forbes=driver.find_elements(By.XPATH,"//div[@class = 'category']")
for i in forbes:
    industry.append(i.text)
    
    


data={"Rank":rank,"Name":name,"Net Worth":net,"Citizenship":country,"Source":source,"Industry":industry}

df=pd.DataFrame(data)

df
```
