```python
DA_Jobs=pd.DataFrame({"Job_Title":Job_Title,"Job_Location":Job_Location,"Company_Name":Company_Name,"Experience_Required":Experience_Required})
DA_Jobs
```

#Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price

```python
#importing the required libraries
import selenium
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome() #connecting to the driver
```

```python
driver.get("https://www.flipkart.com/")#opening the flipkart website on automated edge browser
```

```python
#entering Sunglasses in search bar
Sunglass=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
Sunglass.send_keys("sunglasses")

#clicking Search icon
Search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
Search.click()
```

```python
#Creating the empty lists and storing the data extracted
Brand=[]
Product_Description=[]
Price=[]
Discount=[]

#for searching in the 3 pages
for i in range(3):
    Name=driver.find_elements(By.CLASS_NAME,"_2WkVRV")###Extracting the name of the brand
    for i in Name:
        Brand.append(i.text)
        
        
    Prod=driver.find_elements(By.CLASS_NAME,"IRpwTa")###Extracting the Product description
    for i in Prod:
        Product_Description.append(i.text)
        
        
    Rate=driver.find_elements(By.CLASS_NAME,'_30jeq3')###Extracting price
    for i in Rate:
        Price.append(i.text)
        
    Disc=driver.find_elements(By.CLASS_NAME,'_3Ay6Sb')###Extracting the discount
    for i in Disc:
        Discount.append(i.text)
        
#for letting the python sleep, till the page loads
time.sleep(3)

#going to next page
Next_Page=driver.find_element(By.CLASS_NAME,'_1LKTO3')
Next_Page.click()
```

```python
#Creating the Pandas Data Frame and saving the results in data FRame
list_=list(zip(Brand,Product_Description,Price,Discount))
df4=pd.DataFrame(list_,columns=['Brand','Product_Description','Price','Discount'])
print('100 sunglasses listing on flipkart')
df4.head(100)
```

#Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#importing the required libraries
import selenium
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
#for clicking on top quotes
qu=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
qu.click()
```

```python
Quote=[]  #creating empty list for storing the data
Author=[]
Quotes_Type=[]

#for getting through all 10 pages
for i in range(10):
    Quo=driver.find_elements(By.XPATH,'//a[@class="title"]')###for getting the quote
    for i in Quo:
        Quote.append(i.text)
        
        
    Authors=driver.find_elements(By.XPATH,'//div[@class="author"]')###for getting the Author name
    for i in Authors:
        Author.append(i.text)
        
        
    Type=driver.find_elements(By.XPATH,'//div[@class="tags"]')###for getting the quote type
    for i in Type:
        Quotes_Type.append(i.text)
        
        
##going to next page        
time.sleep(5)
next_page=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]')
next_page.click()
```

```python
#creating the pandas data Frame and calling it
Top1000_Quotes=pd.DataFrame({"Quote":Quote,"Author":Author,"Quotes_Type":Quotes_Type})
Top1000_Quotes
```
